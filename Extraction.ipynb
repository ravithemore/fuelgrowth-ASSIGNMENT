```python
import os
import pandas as pd
import subprocess

# Loading the Dataset
def load_dataset():
    dataset_path = r"C:\Users\cheta\Downloads\Assignment Data - Sheet1.csv"
    df = pd.read_csv(dataset_path)
    print(f"Loaded dataset with {len(df)} rows.")
    return df

data = load_dataset()

#Creating Directory for Videos
os.makedirs("videos", exist_ok=True)

# Downloaded Videos Using yt-dlp
def download_video(url, output_path):
    try:
        command = ["yt-dlp", "-o", output_path, url]
        subprocess.run(command, check=True)
        print(f"Downloaded: {output_path}")
    except subprocess.CalledProcessError as e:
        print(f"Failed to download video from {url}: {e}")

# Iterating Through URLs to Download Videos
for i, row in data.iterrows():
    video_url = row['Video URL']
    output_path = f"videos/{i}.mp4"
    print(f"Downloading video {i} from URL: {video_url}")
    download_video(video_url, output_path)

print("All videos downloaded.")
```
